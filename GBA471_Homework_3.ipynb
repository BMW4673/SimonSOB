```python
#Install Packages
#!pip install pandas xlrd openpyxl

import pandas as pd
import matplotlib.pyplot as plt

#Load Bank.xls into Panda
df=pd.read_excel('BankSalaries-2.xlsx', engine='openpyxl')

#Display the first few rows to verify
print(df.head())

xls=pd.ExcelFile('BankSalaries-2.xlsx', engine='openpyxl')
print(xls.sheet_names)

print('')
#Get basic information about the dataframe
print('\nDataframe Info:')
print(df.describe().round(2))

#Get Basic('\nBasic Statistics:')
```

```python
#Numerical Summaries with Pandas
#Summary statistics for the numerical columns
print(df['Salary'].describe())

print('')
print('GROUPED SUMMARIES')
#Grouped summaries
print(df.groupby('JobGrade')['Salary'].mean())

print('')
print('COUNT OF PEOPLE PER EDUCATION OR JOB GRADE')
#Count of people per Education Level
print(df['Educ.Level'].value_counts())
print(df['JobGrade'].value_counts())

print('')
print('CROSS-TAB BETWEEN GENDER AND JOB GRADE')
#Cross-Tab between Gender and Job Grade
print(pd.crosstab(df['Gender'], df['JobGrade']))
```

```python
#Histogram of Salary
print('')
print('Histogram of Salary')
df['Salary'].hist(bins=30)
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()

#Boxplot of Salary by JobGrade
print('')
print('Boxplot of Salary by JobGrade')
df.boxplot(column='Salary', by='JobGrade')
plt.title('Salary by Job Grade')
plt.suptitle('') #Removes automatic boxplot grouped by label
plt.xlabel('Job Grade')
plt.ylabel('Salary')
plt.show()

#Bar chart: Count of Education Level
print('')
print('Bar Chart of Education Level')
df['Educ.Level'].value_counts().plot(kind='bar')
plt.title('Count by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.show()
```
