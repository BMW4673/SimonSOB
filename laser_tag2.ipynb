```python
# ===============================================================
# Putting It All Together (Laser Tag) (Starter)
# ===============================================================

# ===============================================================
# Part A: Team Roster Dashboard Prototype
# ===============================================================
# PSEUDOCODE:
# 1. Define player rosters for both Red and Green Teams.
# 2. Calculate the number of players on each team.
# 3. Display a formatted list of players for each team with numbering.
# ---------------------------------------------------------------

# Player Names
red_team_players = ["Einstein", "Bond", "Spock", "Noob", "Snake", "Pineapple", "Kirk", "Worf", "Bean", "Mabel"]
green_team_players = ["Cupcake", "Boss Level", "Pudding", "Watermelon", "Ham Sandwich", "Hot Sauce", "Jelly Bean", "Rambo", "Weasley", "Sherlock"]

# Determine team sizes
num_green_players = len(green_team_players)
num_red_players = len(red_team_players)

# Print formatted roster tables
print(
    f"Part A: Team Roster Dashboard Prototype\n",
    f"---------------------------------------\n",
    f"\n"
    f"The Green Team has {num_green_players} players.\n"
    f"\n"
    f"Player\t\tName\n"
    f"------\t\t----"
)
for index, name in enumerate(green_team_players, start=1):
    print(f"{index}\t\t{name}")

print(
    f"\n"
    f"The Red Team has {num_red_players} players.\n"
    f"\n"
    f"Player\t\tName\n"
    f"------\t\t----"
)
for index, name in enumerate(red_team_players, start=1):
    print(f"{index}\t\t{name}")

# ===============================================================
# Part B: Game Summary Dashboard Prototype
# ===============================================================
# PSEUDOCODE:
# 1. Initialize lists for total team scores by game for both teams.
# 2. Loop through each gameâ€™s player scores to calculate total team points.
# 3. Append the calculated totals to team score lists.
# 4. Print formatted output for each game with comma separators.
# ---------------------------------------------------------------

green_team_scores = []
red_team_scores = []

# Calculate total team scores for each game
for game_scores in green_player_scores:
    total = 0
    for score in game_scores:
        total += score
    green_team_scores.append(total)

for game_scores in red_player_scores:
    total = 0
    for score in game_scores:
        total += score
    red_team_scores.append(total)

# Display formatted game summary
print(
    f"Part B: Game Summary Dashboard Prototype A\n",
    f"------------------------------------------\n",
    f"\n"    
    f"Game\t\tGreen\t\tRed\n",
    f"-----\t\t-----\t\t---"
)
for game in range(len(green_team_scores)):
    print(f" {game + 1}\t\t{green_team_scores[game]:,}\t\t{red_team_scores[game]:,}")

# ===============================================================
# Part C: Game Summary Dashboard Prototype B (Enhanced Table)
# ===============================================================
# PSEUDOCODE:
# 1. Import the tabulate package for clean formatted tables.
# 2. For each game:
#    - Retrieve Green and Red team scores.
#    - Compute total points scored in that game.
#    - Determine the winner (Green or Red).
# 3. Print a formatted table with all the above information.
# ---------------------------------------------------------------

from tabulate import tabulate

print(
    f"Part C: Game Summary Dashboard Prototype B\n",
    f"------------------------------------------\n",
    f"\n",
    f"=======================================================================\n",
    f"| Game\t|   Green\t|   Red\t|   Total\t|   Winner\t|\n",
    f"======================================================================="
)

for game in range(len(green_team_scores)):
    green = green_team_scores[game]
    red = red_team_scores[game]
    total = green + red
    winner = "Green" if green > red else "Red"
    print(f"|  {game}\t|   {green:,}\t|   {red:,}\t|   {total:,}\t|   {winner}\t|\t\n",
          "-----------------------------------------------------------------------")

# ===============================================================
# Part D: Determine Game Winners and Totals
# ===============================================================
# PSEUDOCODE:
# 1. Initialize empty lists for game winners and total points per game.
# 2. Loop over all games.
# 3. Calculate total points per game and determine which team won.
# 4. Append the winner ("Green", "Red", or "Tie") and total points.
# ---------------------------------------------------------------

game_winners = []
games_total_points = []

for winners in range(len(green_team_scores)):
    total_points = green_team_scores[winners] + red_team_scores[winners]
    games_total_points.append(total_points)

    if green_team_scores[winners] > red_team_scores[winners]:
        game_winners.append("Green")
    elif green_team_scores[winners] < red_team_scores[winners]:
        game_winners.append("Red")
    else:
        game_winners.append("Tie")

# ===============================================================
# Part E: Green Team Post-Game Dashboard
# ===============================================================
# PSEUDOCODE:
# 1. Initialize variables to track:
#    - Number of games won
#    - Win percentage
#    - Team average score
#    - All-Stars and MVP details
# 2. Loop over the list of game winners and increment wins for Green.
# 3. Calculate win percentage.
# 4. (Later) Extend logic for all-stars and MVP calculations.
# ---------------------------------------------------------------

num_games_won_by_green_team = 0
green_team_win_percentage = 0
green_team_average_score = 0
green_player_total_points = 0
all_stars = []
mvp = 0
mvp_total_points = 0

for winner in game_winners:
    if winner == "Green":
        num_games_won_by_green_team += 1

num_games = len(green_team_scores)
green_team_win_percentage = (num_games_won_by_green_team / num_games) * 100
```
