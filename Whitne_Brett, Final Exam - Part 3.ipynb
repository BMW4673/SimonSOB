```python
# Import Packages
from scipy.stats import poisson

# Define Consants
lam = 9
```

```python
#---------------------------------------------
# Part A - Probability that the system is under high load in any given minute
#---------------------------------------------

# Calculate poisson
p_high_load = 1 - poisson.cdf(13, lam)

# Output statement
print(f"The probability that the system is under high load in any given minute, p(T >= 14), is {p_high_load:.4f} or {p_high_load * 100:.2f}%")
```

```python
#---------------------------------------------
# Part B - Probability that the system will be under high load in a full one-minute interval
#---------------------------------------------
#  Naming convention:
#  p = probability
#  T = t-score
#  ge = greater than or equal to

# Calculate poisson
p_T_ge_14 = 1 - poisson.cdf(13, lam)
p_T_ge_8  = 1 - poisson.cdf( 7, lam)
p_conditional = p_T_ge_14 / p_T_ge_8

# Output statement
print(f"The probability that the system will be under high load in a full one-minute interval, P(T >= 14 | T >= 8), is {p_conditional:,.4f} or {p_conditional * 100:,.2f}%")
```

```python
#---------------------------------------------
# Part C - Probability that the firm exceeds 550 transactions in an hour
#---------------------------------------------

# Define variables
lam_per_hour = lam * 60

# Calculate poisson
p_over_550   = 1 - poisson.cdf(550, lam_per_hour)

# Output statement
print(f"The probability that the firm exceeds 550 transactions in an hour, p(T_hour > 550), is {p_over_550:,.4f} or {p_over_550 * 100:,.2f}%")
```
